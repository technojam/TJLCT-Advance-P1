```python
import pandas as pd
from jupyterthemes import jtplot
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('student-mat.csv')

df
```

```python
df2=pd.read_csv('student-por.csv')
df2
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,cmap='hot')
plt.yticks(rotation=0)
```

The above heatmap doesn't show that only exam grades are correlated to each other

```python

sns.barplot(x=df['Dalc'],y=(df['G1']+df['G2']+df['G3'])/3,hue=df['school'])
plt.xlabel('Daily Consumptions')
plt.ylabel('Average Scores')
```

```python
sns.lineplot(x=df['absences'],y=(df['G1']+df['G2']+df['G3']/3))
plt.xlabel('Number Of School Absences')
plt.ylabel('Average Scores')
```

```python
df['Scores']=(df['G1']+df['G2']+df['G3'])/3
```

```python
sns.lineplot(x='Dalc',y='studytime',data=df)
```

```python
sns.lineplot(x='Dalc',y='failures',data=df)
```

From all the above graphs and chart we came to conclusion that the alcohol consumption of students leading to failures 
as the number of studytime is decreasing and number of failures is increasing
